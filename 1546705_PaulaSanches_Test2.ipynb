```python
# X represents the 2 input features and y represents the 3 classes.
import numpy as np
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=1200, centers=3, n_features=2, center_box=(-4.5, 4.5),
 random_state=21)
print(X.shape)
```

```python
# Plot the generated blobs
plt.figure(1, figsize=(8, 6))
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "ys", label="Yellow Blobs")
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^", label="Green Blobs")
plt.plot(X[:, 0][y==2], X[:, 1][y==2], "rP", label="Red Blobs")
plt.xlabel("Length", fontsize=14)
plt.ylabel("Width", fontsize=14)
plt.legend(loc="upper left", fontsize=12)
```

```python
#logistic Regression

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (
X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(solver ="lbfgs")
log_reg.fit(X_train, y_train)
```

```python
y_pred=log_reg.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))

from sklearn import metrics
print("Confusion Matrix: ", metrics.confusion_matrix(y_test, y_pred))

from sklearn import metrics
print("Classification Report: ", metrics.classification_report(y_test, y_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = .02  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = log_reg.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(8, 6))
plt.pcolormesh(xx, yy, Z, cmap='Blues',)

# Plot the training points
plt.plot(X[:, 0][y==0], X[:, 1][y==0], "ys", label="Yellow Blobs")
plt.plot(X[:, 0][y==1], X[:, 1][y==1], "g^", label="Green Blobs")
plt.plot(X[:, 0][y==2], X[:, 1][y==2], "rP", label="Red Blobs")
plt.xlabel('Length')
plt.ylabel('Width')
plt.legend(loc="upper left", fontsize=12)
plt.show()
```

```python
#SVM

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
svm_clf = Pipeline([
        ("scaler", StandardScaler()),
        ("linear_svc", LinearSVC()),
    ])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42)
```

```python
svm_clf.fit(X_train, y_train)
```

```python
y_pred_svm=svm_clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred_svm))

from sklearn import metrics
print("Confusion Matrix: ", metrics.confusion_matrix(y_test, y_pred_svm))

from sklearn import metrics
print("Classification Report: ", metrics.classification_report(y_test, y_pred_svm))
```
